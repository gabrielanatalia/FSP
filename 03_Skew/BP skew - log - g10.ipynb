# Other Data

```python
fwd = ['EUR1M CMPN Curncy', 'GBP1M CMPN Curncy', 'AUD1M CMPN Curncy', 
       'NZD1M CMPN Curncy', 'JPY1M CMPN Curncy', 'CAD1M CMPN Curncy', 
       'SEK1M CMPN Curncy', 'NOK1M CMPN Curncy', 'CHF1M CMPN Curncy']

fwd_scale_g10 = {"EURUSD": 10000,
             "GBPUSD" : 10000,
             "AUDUSD" : 10000,
             "NZDUSD" : 10000,
             "USDJPY" : 100,
             "USDCAD" : 10000,
             "USDSEK" : 10000,
             "USDNOK" : 10000,
             "USDCHF" : 10000}

fwd_data = new_raw_data[fwd]
fwd_data.columns = fwd_scale_g10.keys()
for c in fwd_scale_g10.keys():
  fwd_data[c] = fwd_data[c]/fwd_scale_g10[c]
  
fwd_price_ori = spot_price_ori[fwd_scale_g10.keys()] + fwd_data
```

```python
bp_usd = fwd_price_ori.pipe(standardize_series, "prices").pipe(np.log10) - spot_price_ori[fwd_scale_g10.keys()].pipe(standardize_series, "prices").pipe(np.log10)
```

# Signal

```python
def custom_skew(series, window):
  num = (series - series.rolling(window).mean())**3
  denom = series.rolling(window).std(ddof=0)**3
  return (num/denom).rolling(window).mean()
```

```python
sig = bp_usd.ffill().pipe(custom_skew, 21*3).pipe(cap_helper, 10).pipe(rolling_mean, 5) * -1
```

```python
bp_usd.ffill().pipe(custom_skew, 21*3).plot()
plt.axhline(-10, linestyle='--', color='red', lw=1, zorder = 2)
plt.axhline(10, linestyle='--', color='red', lw=1, zorder = 2)
```

# Backtest Daily Rebal

```python
daily_bt = Backtest(signal = sig, asset_returns = total_returns, tcost = abs(total_returns*0), settings = settings)
daily_bt.run()
```

```python
get_summary(daily_bt)
```

```python
get_returns_stats(daily_bt.asset_rtn, daily_bt.wts)
```

```python
plot_bt(daily_bt, t="Full Backtest Daily")
```

```python
weekly_bt = Backtest(signal = sig, 
             asset_returns = total_returns, 
             tcost = abs(total_returns*0), 
             settings = settings,
             settings_inject = {"rebalance_period":"friday"})
weekly_bt.run()
```
