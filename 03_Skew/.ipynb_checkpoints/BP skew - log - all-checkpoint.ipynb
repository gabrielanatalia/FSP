# Signal

```python
def custom_skew(series, window):
  num = (series - series.rolling(window).mean())**3
  denom = series.rolling(window).std(ddof=0)**3
  return (num/denom).rolling(window).mean()
```

```python
sig = bp_usd.ffill().pipe(custom_skew, 21*3).pipe(cap_helper, 10).pipe(rolling_mean, 5) * -1
```

```python
bp_usd.ffill().pipe(custom_skew, 21*3).plot()
plt.axhline(-10, linestyle='--', color='red', lw=1, zorder = 2)
plt.axhline(10, linestyle='--', color='red', lw=1, zorder = 2)
```

# Backtest Daily Rebal

```python
daily_bt = Backtest(signal = sig, 
                    asset_returns = total_returns, 
                    tcost = abs(total_returns*0), 
                    settings = settings,
                    asset_groups={"ASIA":ASIA_FX, "G10":G10_FX})
daily_bt.run()
```

```python
get_returns_stats(daily_bt.asset_rtn, daily_bt.wts)
```

```python
plot_bt2(daily_bt, t="Full Backtest Daily")
```

# Sensitivity Analysis

```python
def signal_generation(lookback):
    sig = bp_usd.ffill().pipe(custom_skew, lookback).pipe(cap_helper, 10).pipe(rolling_mean, 5) * -1
    return sig
```

```python
signalsa = [signal_generation(i) for i in [int(252/12),int(252/4),int(252/2),252,252*2,252*3,252*4,252*5]] 
# signalsa
```

```python
backtestsa = [Backtest(signal = x, 
                    asset_returns = total_returns, 
                    tcost = abs(total_returns*0), 
                    settings = settings, 
                    asset_groups={"ASIA":ASIA_FX, "G10":G10_FX}) for x in signalsa]
for testsa in backtestsa:
    testsa.run()
```

```python
sa_variants = pd.DataFrame([x.model_rtn for x  in backtestsa]).T
sa_variants.columns = ["1M lookback", "3M lookback", "6M lookback", "1Y lookback", "2Y lookback","3Y lookback","4Y lookback","5Y lookback"]
gen_plot_rtn(sa_variants)
```
