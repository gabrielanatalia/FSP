# Signal 

```python
def winsorizer(x, width, zcap):
  mu = rolling_mean(x, width, method="simple")
  sigma = rolling_vol(x, width, method="simple")
  z = (x-mu)/sigma
  winsorized = cap_helper(z, zcap) #* sigma + mu
  return winsorized
```

```python
sig = spot_returns.pipe(fill_date, True).pipe(winsorizer, 252, 4).rolling(252*5).skew() + \
        spot_returns.pipe(fill_date, True).pipe(winsorizer, 252, 4).rolling(252).skew()
```

```python
sig = sig.pipe(norm, 252).pipe(demean_xs).pipe(top_bottom_xs, 2).pipe(cap_helper, 2).pipe(rolling_mean, 5) * -1
```

# Backtest Daily Rebal

```python
daily_bt = Backtest(signal = sig, 
                    asset_returns = total_returns, 
                    tcost = abs(total_returns*0), 
                    settings = settings)
daily_bt.run()
```

```python
get_summary(daily_bt)
```

```python
get_returns_stats(daily_bt.asset_rtn, daily_bt.wts)
```

```python
plot_bt(daily_bt)
```

# Backtest Weekly Rebal

```python
weekly_bt = Backtest(signal = sig, 
             asset_returns = total_returns, 
             tcost = abs(total_returns*0), 
             settings = settings,
             settings_inject = {"rebalance_period":"friday"})
weekly_bt.run()
```

```python
get_summary(weekly_bt)
```

```python
plot_bt(weekly_bt)
```
