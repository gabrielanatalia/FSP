```python
import sys
sys.path.insert(1, '../')
from genesis import *
```

```python
universe = G10_FX
total_returns = total_returns_usd[universe]
spot_returns = spot_returns_usd[universe]
carry_returns = carry_returns_usd[universe]
```

# Settings

```python
settings = Settings(start_date = "2000-01-01", end_date = "2016-12-31", notional = 100, implementation_lag = 1,
             asset_scaling_method="simple", asset_scaling_target=None, asset_scaling_width=130,
             sector_scaling_method="simple", sector_scaling_target=0.1, sector_scaling_width=130,
             portfolio_scaling_method="simple", portfolio_scaling_target=0.1, portfolio_scaling_width=260)
```

# Signal

```python
def afunc(x):
  if pd.isna(x):
    return np.nan
  elif x < -1:
    return 1
  elif x > 1:
    return -1
  return 0
```

```python
# difference = spot_returns.pipe(demean, window=252*5, min_periods=252*3)
# deviations = difference.expanding(min_periods = 252).std(ddof=0)
# sig = (difference/deviations).applymap(afunc) * (1/rolling_vol(spot_returns, 252)) 
```

```python
spot_rtn_monthly = spot_price_usd[universe].groupby(pd.Grouper(freq='M')).agg("last").pct_change()
difference = spot_rtn_monthly.pipe(demean, window=12*5, min_periods=12*3)
deviations = difference.expanding(min_periods = 12).std(ddof=0)
sig = (difference/deviations).applymap(afunc) * (1/rolling_vol(spot_returns, 12))
sig = sig.pipe(fill_date, True)
```

# Backtest

```python
bt = Backtest(signal = sig, asset_returns = total_returns, 
                     tcost = abs(total_returns*0), 
                     settings = settings,
                     settings_inject={"rebalance_period":"daily"})
bt.run()
```
