1st strategy for carry is to use the spread 2Y and spread 5Y of swaps and take zscore of 1Y  
then take average between the 2 signals

```python
import pandas as pd
import pyfolio as pf
import matplotlib.pyplot as plt
from genesis import *
```

# Price Data 

```python
px = read_prices("../Dymon/Code Data/carry_adj_fx_returns.csv")
px.columns = [x[0] for x in px.columns]
rtns = px.pct_change()

rtns_copy = copy(rtns)

rtns = pd.DataFrame()
for i in list(rtns_copy):
  if i[:3] != "USD":
    col = rtns_copy.loc[:,i] * -1
    col.name = i[3:] + i[:3]
  else:
    col = rtns_copy.loc[:,i]
    col.name = i
  rtns = pd.concat([rtns, col], axis = 1)
  
  
# rtns = px.pct_change()
rtns.index = pd.to_datetime(rtns.index)
rtns = rtns.loc[:"2016-12-31"]
```

```python
# rtns = rtns[["USDAUD", "USDCAD", "USDEUR", "USDGBP", "USDJPY", "USDNZD", "USDSEK", "USDNOK", "USDCHF"]]
rtns = rtns[[x for x in list(rtns) if x not in ["USDAUD", "USDCAD", "USDEUR", "USDGBP", "USDJPY", "USDNZD", "USDSEK", "USDNOK", "USDCHF"]]]
```

# Settings

```python
s = Settings(start_date = "2000-01-01", end_date = "2016-12-31", notional = 100, implementation_lag = 1,
             asset_scaling_method="simple", asset_scaling_target=0.1, asset_scaling_width=130,
             sector_scaling_method="simple", sector_scaling_target=0.1, sector_scaling_width=130,
             portfolio_scaling_method="simple", portfolio_scaling_target=0.1, portfolio_scaling_width=260)
```

# Other Data

```python
swaps = pd.read_csv("../Dymon/Code Data/swaps.csv", index_col=0).iloc[2:]
```

```python
swaps.columns = ["2Y", "5Y"]
swaps = swaps.astype(float)
```

```python
spread_2y = swaps["2Y"]
spread_5y = swaps["5Y"]
spread_2y.index.name = "date"
spread_5y.index.name = "date"
```

```python
corp_spread = pd.read_csv("../Dymon/Code Data/spread.csv", index_col=0).iloc[6:]
```

```python
ig_spread = corp_spread["LGCPOAS Index"].astype(float)
hy_spread = corp_spread["LG50OAS Index"].astype(float)
```

```python
rtns.mean(1).cumsum().plot()
```

```python
spread_2y.plot()
```

# Signal 

```python
signal1 = cap_helper(zscore(spread_5y.ffill(), 252), 2)
signal2 = cap_helper(zscore(spread_2y.ffill(), 252), 2)
signal = (signal1 + signal2)/2
signal = signal2
signal = cap_helper(zscore(hy_spread.ffill(), 252), 2).loc["2010":]
# signal = demean(spread.ffill(), 252)
signal.plot()
sig = pd.concat([signal] * len(rtns.columns), axis=1)
sig.columns = rtns.columns
sig.index = pd.to_datetime(sig.index)
sig = sig.loc[:"2016"]
```

```python
b = Backtest(signal = sig, asset_returns = rtns, tcost = abs(rtns*0), settings = s)
b.run()
```
