# Settings

```python
s = Settings(start_date = "2000-01-01", end_date = "2016-12-31", notional = 100, implementation_lag = 1,
             asset_scaling_method="simple", asset_scaling_target=0.1, asset_scaling_width=130, rebalance_period="daily",
             sector_scaling_method="simple", sector_scaling_target=0.1, sector_scaling_width=130,
             portfolio_scaling_method="simple", portfolio_scaling_target=0.1, portfolio_scaling_width=260)
```

# Other Data

```python
vol_1y = read_prices("../../Dymon/Code Data/vol_12m.csv")
vol_1y.columns = [x[0] for x in vol_1y.columns]
g10_vol = vol_1y[['EURUSDV1Y BGN Curncy', 'GBPUSDV1Y BGN Curncy', 'AUDUSDV1Y BGN Curncy',
                  'NZDUSDV1Y BGN Curncy', 'USDJPYV1Y BGN Curncy', 'USDCADV1Y BGN Curncy',
                  'USDCHFV1Y BGN Curncy', 'USDNOKV1Y BGN Curncy', 'USDSEKV1Y BGN Curncy']]
g10_vol.columns = ["USDEUR", "USDGBP", "USDAUD", 
                   "USDNZD", "USDJPY", "USDCAD",
                   "USDCHF", "USDNOK", "USDSEK"]
g10_vol = g10_vol[carry_returns.columns]
```

```python
g10_vol = total_returns.pipe(fill_date, fill_forward=True).pipe(rolling_mean, method="exponential", window=126)[carry_returns.columns]
```

```python
# sig = (carry_returns.pipe(fill_date, fill_forward=True)/g10_vol.pipe(fill_date, fill_forward=True).pipe(rolling_mean, method="exponential", window=252)).pipe(omit_trailing_na)

sig = (carry_returns.pipe(fill_date, fill_forward=True)/g10_vol)
# sig = sig.pipe(demean_xs)
sig = sig/sig.pipe(rolling_vol, 252)
sig = sig.apply(off_xs, axis=1) * -1
```

# Signal 

```python
b = Backtest(signal = sig, asset_returns = total_returns, tcost = abs(total_returns*0), settings = s)
b.run()
```

```python
bt_obj = b
```
