```python
def unmelt_cftc(country, field):
  country_df = cftc_ori[cftc_ori["currency"]==country]
  country_unmelt = country_df.drop(["currency"], axis=1).pivot(index='date', columns=['data_type'])["position"]
  country_unmelt.columns = list(country_unmelt)
  country_unmelt.index = pd.to_datetime(country_unmelt.index)
  country_unmelt = country_unmelt * cftc_dict[country]
  return country_unmelt[field]
```

```python
am_pos_df = pd.concat([unmelt_cftc(x, "asset_managers_combined_position") for x in g7], axis=1).set_axis(g7, axis=1).rename(columns = lambda x: "USD"+x)
lf_pos_df = pd.concat([unmelt_cftc(x, "leveraged_funds_combined_position") for x in g7], axis=1).set_axis(g7, axis=1).rename(columns = lambda x: "USD"+x)
interest_df = pd.concat([unmelt_cftc(x, "net_combined_open_interest") for x in g7], axis=1).set_axis(g7, axis=1).rename(columns = lambda x: "USD"+x)
comb_pos_df = pd.concat([unmelt_cftc(x, "net_combined_position") for x in g7], axis=1).set_axis(g7, axis=1).rename(columns = lambda x: "USD"+x)
futures_pos_df = pd.concat([unmelt_cftc(x, "net_futures_position") for x in g7], axis=1).set_axis(g7, axis=1).rename(columns = lambda x: "USD"+x)
```

# Signal 

```python
futures_pos_df["USDEUR"].diff().pipe(zscore,52*3).plot()
```

```python
raw_sig = futures_pos_df.pipe(zscore, 52*3) * -1
sig = raw_sig.pipe(fill_date, True)
sig["USDNOK"] = np.nan
sig["USDSEK"] = np.nan
sig = CAP(sig, 2.5)
sig[abs(sig) < 0.25] = 0
sig = sig.pipe(rolling_mean, 5)
```

```python
raw_sig = futures_pos_df.diff().pipe(norm,52*3)
sig = raw_sig.pipe(fill_date, True)
sig["USDNOK"] = np.nan
sig["USDSEK"] = np.nan
sig = CAP(sig, 2.5)
sig[abs(sig) < 0.25] = 0
sig = sig.pipe(rolling_mean, 5)
```

```python
def plot_bt(bt_obj):
  gs = GridSpec(nrows = 4, ncols = 3)
  fig = plt.figure(figsize=(30, 30))
  fig.suptitle(t="Full Backtest", fontsize=30, y=.925)

  fig.add_subplot(gs[0]).set_label(gen_plot_rtn(bt_obj.model_rtn, main = "Returns (Full Period)"))
  fig.add_subplot(gs[1]).set_label(gen_plot_rtn(bt_obj.model_rtn["2012":], main = "Returns (2012/)"))
  fig.add_subplot(gs[2]).set_label(gen_rtn_asset(bt_obj))

  fig.add_subplot(gs[3]).set_label(pf.plot_drawdown_underwater(bt_obj.model_rtn))
  fig.add_subplot(gs[4]).set_label(gen_rtn_component(bt_obj, spot_returns.loc[:"2016-12-31"], carry_returns.loc[:"2016-12-31"]))
  fig.add_subplot(gs[5]).set_label(gen_rtn_longshort(bt_obj))

  fig.add_subplot(gs[6]).set_label(gen_leadlag(bt_obj, nlag=10))
  fig.add_subplot(gs[7]).set_label(gen_rolling_winloss(bt_obj.model_rtn))
  fig.add_subplot(gs[8]).set_label(gen_rolling_hitrate(bt_obj.model_rtn))

  fig.add_subplot(gs[9]).set_label(pf.plot_annual_returns(bt_obj.model_rtn))
  fig.add_subplot(gs[10:]).set_label(gen_signal(bt_obj))
```

```python
plot_bt(bt_obj)
```
