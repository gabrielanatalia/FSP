# Settings

```python
settings = Settings(start_date = "2000-01-01", end_date = "2016-12-31", notional = 100, implementation_lag = 1,
             asset_scaling_method="simple", asset_scaling_target=0.1, asset_scaling_width=130,
             sector_scaling_method="simple", sector_scaling_target=0.1, sector_scaling_width=130,
             portfolio_scaling_method="simple", portfolio_scaling_target=0.1, portfolio_scaling_width=260)
```

# Other Data

```python
swaps = pd.read_csv("../../Dymon/Code Data/swaps.csv", index_col=0).iloc[2:]
```

```python
swaps.columns = ["2Y", "5Y"]
swaps = swaps.astype(float)
swaps.index = pd.to_datetime(swaps.index)
```

```python
spread_2y = swaps["2Y"]
spread_5y = swaps["5Y"]
spread_2y.index.name = "date"
spread_5y.index.name = "date"
```

# Signal

```python
spread_5y.pipe(fill_date, True).pipe(omit_trailing_na).pipe(zscore, 252).pipe(cap_helper, 2).plot()
```

```python
signal1 = spread_5y.pipe(fill_date, True).pipe(omit_trailing_na).pipe(zscore, 252).pipe(cap_helper, 2)
signal2 = spread_2y.pipe(fill_date, True).pipe(omit_trailing_na).pipe(zscore, 252).pipe(cap_helper, 2)
signal = (signal1 + signal2)/2
signal = signal.pipe(rolling_mean, 5)
# signal.plot()
sig = pd.concat([signal] * len(total_returns.columns), axis=1)
sig.columns = total_returns.columns
sig.index = pd.to_datetime(sig.index)
```

```python
b = Backtest(signal = sig, asset_returns = total_returns, tcost = abs(total_returns*0), settings = settings)
b.run()
```

```python
#regime dataframe

def regime_df(returns):
    returns_regime = pd.DataFrame(returns)
    returns_regime.rename(columns={0:'return'}, inplace=True)
    
    returns_regime['cum_return'] = returns_regime['return'].cumsum() #cumulative returns

    #determine market regimes
    returns_regime['regime_class_1'] = np.random.choice(['a','b'], returns_regime.shape[0])
    returns_regime['regime_class_2'] = np.random.choice(['c','d'], returns_regime.shape[0])
    returns_regime["regime_class_comb"] = returns_regime["regime_class_1"] + returns_regime["regime_class_2"]

    def regime_category(row):
        if row['regime_class_comb'] == 'ac':
            value = 0
        elif row['regime_class_comb'] == 'bc':
            value = 1
        elif row['regime_class_comb'] == 'ad':
            value = 2
        else:
            value = 3
        return value
        
    returns_regime['reg_category'] = returns_regime.apply(regime_category, axis=1)
    return(returns_regime)
```

```python
regime_df(bt_obj.model_rtn)
```

```python
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
import matplotlib.dates as mdates
from matplotlib.lines import Line2D

#plot returns
def plot_regime_all(returns):
    returns_regime = regime_df(returns)
    x = returns_regime.index
    y = returns_regime['cum_return']
    
    fig, ax = plt.subplots(figsize = (10,6))

    inxval = mdates.date2num(x) #convert dates to numbers first

    num_classes = 4
    cmap = ListedColormap(['green', 'blue', 'red', 'yellow'])
    norm = BoundaryNorm(range(num_classes+1), cmap.N)
    points = np.array([inxval, y]).T.reshape(-1,1,2)
    segments = np.concatenate([points[:-1],points[1:]], axis=1)

    lc = LineCollection(segments, cmap=cmap, norm=norm)
    lc.set_array(returns_regime['reg_category'])
    lc.set_linewidth(2)
    ax.add_collection(lc)

    
    loc = mdates.AutoDateLocator() #covert numbers back to date
    ax.xaxis.set_major_locator(loc)
    ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(loc)) 
    ax.autoscale_view()

    #add legend and labels
    ax.set_xlabel('Dates')
    ax.set_ylabel('Cumulative returns')
    line_green = Line2D([0], [0], label='Regime 0', color='green')
    line_blue = Line2D([0], [0], label='Regime 1', color='blue')
    line_red = Line2D([0], [0], label='Regime 2', color='red')
    line_yellow = Line2D([0], [0], label='Regime 3', color='yellow')
    plt.legend(handles=[line_green,line_blue,line_red,line_yellow],loc="upper left")

    fig.suptitle('Returns by Market Regime')
    
    ax.grid(linestyle = 'dashed', linewidth = 1, zorder = 1)
    ax.axhline(0.0, linestyle='--', color='red', lw=1, zorder = 2)

    return(plt.show())
```

```python
plot_regime_all(bt_obj.model_rtn)
```

```python
#performance under each regime
def plot_regime(regime_number, returns):
    df = regime_df(returns)
    df['cum_return_reg'] = (df['return']*(df['reg_category']==regime_number)).cumsum()
    
    if regime_number == 0:
        color='green'
    elif regime_number == 1:
        color='blue'
    elif regime_number == 2:
        color='red'
    else:
        color='yellow'
        
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.plot(df.index, df['cum_return_reg'],color=color,label='Regime '+str(regime_number))

    ax.set_xlabel('Date')
    ax.set_ylabel('Cumulative Returns')
    ax.set_title('Returns (Regime '+str(regime_number)+')')
    ax.legend(loc='upper left')
    
    ax.grid(linestyle = 'dashed', linewidth = 1, zorder = 1)
    ax.axhline(0.0, linestyle='--', color='red', lw=1, zorder = 2)
```

```python
plot_regime(0,bt_obj.model_rtn)
plot_regime(1,bt_obj.model_rtn)
plot_regime(2,bt_obj.model_rtn)
plot_regime(3,bt_obj.model_rtn)
```
